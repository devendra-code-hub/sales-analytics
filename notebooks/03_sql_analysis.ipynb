```python
# notebooks/03_sql_analysis.ipynb
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("sqlite:///../database/sales.db")

region_df = pd.read_sql("""
    SELECT Region,
           ROUND(SUM(Sales),2) AS total_sales,
           ROUND(SUM(Profit),2) AS total_profit
    FROM sales GROUP BY Region ORDER BY total_sales DESC
""", engine)

print(region_df)
```
